```python
import json
import cv2
import imageio
import os

# Load JSON file
json_path = "../classification/cycle_splits/labeled_data"
video_path = "C:/awilde/britta/LTU/SkiingProject/SkiersProject/Data/selectedData"

id = "33"
json_filename = json_path + "/" + "labeled_cycles_" + id + ".json"
video_filename = video_path + "/" +"DJI_00" + id + ".mp4"
output_dir = "cycles_gifs"

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Read JSON data
with open(json_filename, "r") as file:
    data = json.load(file)

# Open video file
cap = cv2.VideoCapture(video_filename)

if not cap.isOpened():
    print("Error: Could not open video file.")
    exit()

fps = cap.get(cv2.CAP_PROP_FPS)

# Process each cycle
for i, cycle_key in enumerate(data):
    cycle = data[cycle_key]
    start_frame = cycle["Start_frame"]
    end_frame = cycle["End_frame"]
    label = cycle.get("Label", "Not known") 
    gif_filename = os.path.join(output_dir, f"{id}_cycle_{i + 1}.gif")
    
    frames = []
    cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame)
    
    for frame_num in range(start_frame, end_frame + 1):
        ret, frame = cap.read()
        if not ret:
            print(f"Warning: Could not read frame {frame_num}.")
            break
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)  # Convert to RGB
        
        # Add label text to frame
        font = cv2.FONT_HERSHEY_SIMPLEX
        font_scale = 1
        font_color = (255, 255, 255)  # White text
        thickness = 2
        position = (50, 50)  # Position of text (x, y)
        
        cv2.putText(frame, label, position, font, font_scale, font_color, thickness, cv2.LINE_AA)
        
        frames.append(frame)
    
    # Save frames as GIF
    if frames:
        imageio.mimsave(gif_filename, frames, duration=1000 / fps)
        print(f"Saved: {gif_filename}")
    else:
        print(f"Error: No frames captured for cycle {i + 1}")

# Release video capture
cap.release()
print("Processing complete.")
```
