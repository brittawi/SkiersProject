```python
import matplotlib.pyplot as plt
from utils import load_json, normalize_signal, smooth_signal, compute_relative_keypoints
```

```python

# Function to plot cycles for a specific joint
def plot_cycles(data, joint_name, axis1, axis2):
    plt.figure(figsize=(8, 6))
    
    for key in data.keys():
        cycle_dict = data[key]
        if joint_name + "_x" in list(cycle_dict.keys()):
            axis1_values = cycle_dict[joint_name + "_x"]
            axis2_values = cycle_dict[joint_name + "_y"]
            axis1_values_normalized = normalize_signal(axis1_values)
            axis2_values_normalized = normalize_signal(axis2_values)
            axis1_values_smoothed = smooth_signal(axis1_values_normalized)
            axis2_values_smoothed = smooth_signal(axis2_values_normalized)
            plt.plot(axis1_values_smoothed, axis2_values_smoothed, color='black', alpha=0.5)    
              
    plt.xlabel(f'{axis1} position')
    plt.ylabel(f'{axis2} position')
    plt.title(f'Overlay of {joint_name} movements')
    plt.grid(True)
    plt.show()    
    
```

```python
# Load the data and plot
data_path = './labeled_data/labeled_cycles_11.json'  # Update path if necessary
data = load_json(data_path)
```

```python
joints_to_plot = ["LWrist", "RWrist", "LKnee", "RKnee", "LAnkle", "RAnkle"]

for joint in joints_to_plot:
    axis1, axis2 = 'x', 'y'  # Adjust based on available data
    plot_cycles(data, joint, axis1, axis2)
```
