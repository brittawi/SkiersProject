```python
import sys
import os
from pathlib import Path
# Add the utils directory to sys.path
sys.path.append(str(Path(os.getcwd()).resolve().parents[0]))
print(sys.path)
import torch
from utils.config import update_config 
from alphapose.models import builder

cfg = update_config("../alphapose/configs/256x192_res50_lr1e-3_1x.yaml")
model = builder.build_sppe(cfg.MODEL, preset_cfg=cfg.DATA_PRESET)

if cfg.MODEL.PRETRAINED:
    model_path = os.path.join("../alphapose", cfg.MODEL.PRETRAINED)
    print(model_path)
    model.load_state_dict(torch.load(model_path))
        

model.load_state_dict(torch.load("../alphapose/pretrained_models/RegLoss100EpochNoFlipDPG.pth", map_location="cpu"))  # Load state_dict safely
model.eval()
```

```python
for name, param in model.named_parameters():
    print(f"Layer: {name}")
    print(f"Mean: {param.data.mean()}, Std: {param.data.std()}, Min: {param.data.min()}, Max: {param.data.max()}")
    print("-" * 50)
```
