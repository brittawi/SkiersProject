# Convert annotations from halpe to coco format
This script can be used to convert annotations from the halpe format that we get when running alphapose to the coco format, which is needed for CVAT and the fine-tuning process.

For this the folder with the annotations has to be specified as well as the output folder.

```python
from utils.annotation_format import halpe26_to_coco
import os
import glob
import json

#PATH_ANNO = "C:\\Users\\britt\OneDrive - Luleå University of Technology\\SkiersProject\\Code\\resultsAlphapose\\Json"
PATH_ANNO = "C:\\Users\\britt\\AlphaPose\\examples\\res\\"
OUTPUT_FOLDER = "C:\\Users\\britt\OneDrive - Luleå University of Technology\\SkiersProject\\Code\\output_halpe_to_coco_finetuned"
# create folder if it does not exist
if not os.path.exists(OUTPUT_FOLDER):
   os.makedirs(OUTPUT_FOLDER)
   
annotation_files = glob.glob(PATH_ANNO + "/*.json")
print(len(annotation_files))

for anno in annotation_files:
    filename = os.path.basename(anno).split(".")[0] + "_coco.json"
    output_path = os.path.join(OUTPUT_FOLDER, filename)
    coco_data = halpe26_to_coco(anno, output_path)
   # Save the updated COCO JSON file
    with open(output_path, 'w') as f:
        json.dump(coco_data, f, indent=4)

```
